```python
import pandas as pd

df_a = pd.read_csv('data_by_A.csv')
df_a
```

```python
df_b = pd.read_csv('data_by_B.csv')
df_b
```

```python
df_total = pd.concat([df_a, df_b], axis=0, ignore_index=True).drop('Unnamed: 0', axis=1)
df_total
```

## 데이터 살펴보기

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
plt.scatter(df_total['branch_num'], df_total['sales_total'])
plt.show()
```

```python
df_total.dropna(inplace=True)
df_total.tail()
```

```python
df_total.describe()
```

## 데이터셋 분류하기

```python
np_total = df_total.to_numpy()
np_total
```

```python
data_input = np_total[:,0]
data_input
```

```python
data_label = np_total[:,1]
data_label
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_label, test_label = train_test_split(data_input, data_label)
```

## 모델링

```python
train_input_reshaped = train_input.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(train_input_reshaped, train_label)
```

## 검증

```python
lr.coef_
```

```python
lr.intercept_
```

```python
def linear_equation(x):
    return (lr.coef_ * x) +lr.intercept_
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
plt.scatter(df_total['branch_num'], df_total['sales_total'])
plt.plot(
    [25,60], 
    [linear_equation(25),linear_equation(60)]
)
plt.show()
```

```python
test_input_reshaped = test_input.reshape(-1,1)
test_input_reshaped
```

```python
lr.score(test_input_reshaped, test_label)
```

```python
lr.score(train_input_reshaped, train_label)
```

## 특성공학

```python
type(data_input)
```

```python
data_input_squared = data_input * data_input
data_input_squared
```

```python
data_input_new = list(zip(data_input_squared, data_input))
data_input_new
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_label, test_label = train_test_split(data_input_new, data_label)
```

```python
from sklearn.linear_model import LinearRegression

lr_new = LinearRegression()
lr_new.fit(train_input, train_label)
```

## 검증하기

```python
import numpy as np
import matplotlib.pyplot as plt



inp = int(input())
prediction = lr_new.predict(
    [[inp**2, inp]]
)

plt.figure(figsize=(10,8))
plt.scatter(df_total['branch_num'], df_total['sales_total'])

point = np.arange(25,inp)

plt.plot(
    point,
    (
        (point**2)*lr_new.coef_[0]
        + point*lr_new.coef_[1]
        + lr_new.intercept_
    )
)

plt.scatter(inp, prediction)

plt.show()
```
